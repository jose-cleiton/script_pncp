```python
import re
import pandas as pd

# =========================
# AJUDAS PARA REGEX FLEXÍVEL
# =========================

SEP = r"[\s\-_/]*"   # separador flexível: espaço, hífen, underscore, barra

def flex_term(termo: str) -> str:
    """
    Converte uma frase simples em regex flexível entre palavras.
    Ex.: 'controle de acesso' -> 'controle[\\s\\-_/]*de[\\s\\-_/]*acesso'
    """
    partes = termo.strip().split()
    return SEP.join(map(re.escape, partes))

# =========================
# PADRÕES POSITIVOS FLEXÍVEIS
# =========================
PADROES_POSITIVOS = {
    "software_controle_acesso": rf"\b{flex_term('software de controle de acesso')}\b",
    "software_seguranca": r"\bsoftware" + SEP + r"de" + SEP + r"seguran[cç]a\b",
    "software_monitoramento": rf"\b{flex_term('software de monitoramento')}\b",
    "software_portaria": rf"\b{flex_term('software para controle de portaria')}\b",
    "software_registro_frequencia": r"\bsoftware" + SEP + r"para" + SEP + r"registro" + SEP + r"de" + SEP + r"frequ[eê]ncia\b",
    "software_gestao": r"\bsoftware" + SEP + r"de" + SEP + r"gest[aã]o\b",

    "relogio_ponto": r"\brel[oó]gio" + SEP + r"de" + SEP + r"ponto\b",
    "cartao_ponto": r"\bcart[aã]o" + SEP + r"de" + SEP + r"ponto\b",
    "controle_ponto": rf"\b{flex_term('controle de ponto')}\b",

    "reconhecimento_facial": rf"\b{flex_term('reconhecimento facial')}\b",
    "registrador_eletronico_ponto_biometrico": (
        r"\bregistrador" + SEP + r"eletr[oô]nico" + SEP + r"de" + SEP + r"ponto" + SEP + r"biom[eé]trico\b"
    ),
    "gerenciamento_eletronico_frequencia": (
        r"\bgerenciamento" + SEP + r"eletr[oô]nico" + SEP + r"de" + SEP + r"frequ[eê]ncia\b"
    ),
    "controle_frequencia": r"\bcontrole" + SEP + r"de" + SEP + r"frequ[eê]ncia\b",
    "registrador_ponto_facial": r"\bregistrador" + SEP + r"de" + SEP + r"ponto" + SEP + r"facial\b",
    "sistema_aplicativo_facial": r"\bsistema" + SEP + r"com" + SEP + r"aplicativo" + SEP + r"facial\b",
    "controle_frequencia_facial": r"\bcontrole" + SEP + r"de" + SEP + r"frequ[eê]ncia" + SEP + r"facial\b",
    "controle_faces": rf"\b{flex_term('controle de faces')}\b",
    "biometria_face": r"\bequipamento" + SEP + r"com" + SEP + r"biometria" + SEP + r"de" + SEP + r"face\b",
    "identificador_facial": rf"\b{flex_term('identificador facial')}\b",

    "bastao_ronda": r"\bbast[aã]o" + SEP + r"de" + SEP + r"ronda\b",
    "relogio_protocolador": r"\brel[oó]gio" + SEP + r"protocolador\b",
    "relogio_cartografico": r"\brel[oó]gio" + SEP + r"cartogr[aá]fico\b",
    "bobina_termica_relogio_ponto": (
        r"\bbobina" + SEP + r"t[eé]rmica" + SEP + r"para" + SEP + r"rel[oó]gio" + SEP + r"de" + SEP + r"ponto\b"
    ),
    "papel_termosensivel_relogio_ponto": (
        r"\bpapel" + SEP + r"termosens[ií]vel" + SEP + r"para" + SEP + r"rel[oó]gio" + SEP + r"de" + SEP + r"ponto\b"
    ),

    "controlador_acesso": rf"\b{flex_term('controlador de acesso')}\b",
    "fechadura_facial": rf"\b{flex_term('fechadura facial')}\b",
    "fechadura_biometrica": r"\bfechadura" + SEP + r"biom[eé]trica\b",
    "fechadura_eletronica_facial": r"\bfechadura" + SEP + r"eletr[oô]nica" + SEP + r"facial\b",
    "fechadura_eletromagnetica": r"\bfechadura" + SEP + r"eletromagn[eé]tica\b",
    "controladora_acesso_facial": r"\bcontroladora" + SEP + r"de" + SEP + r"acesso" + SEP + r"facial\b",

    "hid": r"\bhid\b",
    "hikvision": r"\bh[ií]?kvision\b",   # pega Hikvision e variação digitada errada simples
    "detector_metais": rf"\b{flex_term('detector de metais')}\b",
    "porta_giratoria": r"\bporta" + SEP + r"girat[oó]ria\b",
    "portal_detector_metal": r"\bportal" + SEP + r"detector" + SEP + r"metal\b",
    "porta_automatica": r"\bporta" + SEP + r"autom[aá]tica\b",
    "motor_portao": r"\bmotor" + SEP + r"de" + SEP + r"port[aã]o\b",

    "catraca_facial": r"\bcatraca(?:s)?" + SEP + r"facial(?:is)?\b",
    "catraca_balcao": r"\bcatraca(?:s)?" + SEP + r"balc[aã]o\b",
    "catraca_mecanica": r"\bcatraca(?:s)?" + SEP + r"mec[aâ]nica(?:s)?\b",
    "catraca_eletronica": r"\bcatraca(?:s)?" + SEP + r"eletr[oô]nica(?:s)?\b",
    "catraca_pedestal_bidirecional": r"\bcatraca(?:s)?" + SEP + r"pedestal" + SEP + r"bidirecional\b",
    "catraca_pne": r"\bcatraca(?:s)?" + SEP + r"pne\b",

    "controle_acesso": rf"\b{flex_term('controle de acesso')}\b",
    "controle_acesso_facial": r"\bcontrole" + SEP + r"de" + SEP + r"acesso" + SEP + r"facial\b",
    "cancela_automatica": r"\bcancela(?:s)?" + SEP + r"autom[aá]tica(?:s)?\b",
    "cancela_eletronica": r"\bcancela(?:s)?" + SEP + r"eletr[oô]nica(?:s)?\b",
    "leitor_facial": r"\bleitor(?:es)?" + SEP + r"facial(?:is)?\b",
    "torniquete": r"\btorniquete(?:s)?\b",
    "conjunto_controle_area_restrita": (
        r"\bconjunto" + SEP + r"controle" + SEP + r"acesso" + SEP + r"[aá]rea" + SEP + r"restrita\b"
    ),
    "idface": r"\bidface\b",
    "botoeira": r"\bbotoeira(?:s)?\b",
}

# =========================
# PADRÕES NEGATIVOS FLEXÍVEIS
# =========================
PADROES_NEGATIVOS = {
    "medicamentos": r"\bmedicamento(?:s)?\b",
    "cadeiras_escolares": r"\bcadeira(?:s)?" + SEP + r"escolar(?:es)?\b",
    "merenda_escolar": rf"\b{flex_term('merenda escolar')}\b",
    "generos_alimenticios": r"\bg[eê]neros?" + SEP + r"aliment[ií]cios?\b",
    "material_limpeza": rf"\b{flex_term('material de limpeza')}\b",
    "combustiveis": r"\bcombust[ií]vel(?:eis|es)?\b",
    "frota": r"\bfrota(?:s)?\b",
    "rastreamento_veicular": r"\brastreamento" + SEP + r"veicular\b",
    "gps": r"\bgps\b",
    "telemetria": r"\btelemetria\b",
    "uniforme": r"\buniforme(?:s)?\b",
    "livros_didaticos": r"\blivro(?:s)?" + SEP + r"did[aá]tico(?:s)?\b",
    "software_multas": r"\bsoftware" + SEP + r"de" + SEP + r"multa(?:s)?\b",
    "prontuario_eletronico": r"\bprontu[aá]rio" + SEP + r"eletr[oô]nico\b",
    "erp_hospitalar": r"\berp" + SEP + r"hospitalar\b",
    "buffet": r"\bbuffet\b",
    "estacao_solda": r"\besta[cç][aã]o" + SEP + r"de" + SEP + r"solda\b",
    "solda": r"\bsolda\b",
}

REGEX_POSITIVOS = {k: re.compile(v, flags=re.IGNORECASE) for k, v in PADROES_POSITIVOS.items()}
REGEX_NEGATIVOS = {k: re.compile(v, flags=re.IGNORECASE) for k, v in PADROES_NEGATIVOS.items()}

# =========================
# LIMPEZA
# =========================
def limpar_texto(texto: str) -> str:
    if pd.isna(texto):
        return ""
    texto = str(texto).strip().lower()
    texto = re.sub(r"\s+", " ", texto)
    return texto

# =========================
# CLASSIFICAÇÃO
# =========================
def classificar_regex(texto: str):
    t = limpar_texto(texto)

    positivos = [nome for nome, rgx in REGEX_POSITIVOS.items() if rgx.search(t)]
    negativos = [nome for nome, rgx in REGEX_NEGATIVOS.items() if rgx.search(t)]

    if positivos and not negativos:
        return {
            "classificacao_final": 1,
            "status_final": "relevante",
            "termos_positivos_batidos": ", ".join(positivos[:10]),
            "termos_negativos_batidos": "",
            "motivo": f"Padrões positivos encontrados: {', '.join(positivos[:5])}"
        }

    if negativos and not positivos:
        return {
            "classificacao_final": 0,
            "status_final": "nao_relevante",
            "termos_positivos_batidos": "",
            "termos_negativos_batidos": ", ".join(negativos[:10]),
            "motivo": f"Padrões negativos encontrados: {', '.join(negativos[:5])}"
        }

    if positivos and negativos:
        return {
            "classificacao_final": "",
            "status_final": "revisar",
            "termos_positivos_batidos": ", ".join(positivos[:10]),
            "termos_negativos_batidos": ", ".join(negativos[:10]),
            "motivo": "Conflito entre padrões positivos e negativos"
        }

    return {
        "classificacao_final": "",
        "status_final": "revisar",
        "termos_positivos_batidos": "",
        "termos_negativos_batidos": "",
        "motivo": "Nenhum padrão forte encontrado"
    }

def classificar_lista(textos):
    resultados = []
    for texto in textos:
        r = classificar_regex(texto)
        resultados.append({
            "texto": texto,
            **r
        })
    return pd.DataFrame(resultados)

# =========================
# EXEMPLO
# =========================
if __name__ == "__main__":
    textos = [
        "LOCAÇÃO DE CATRACAS-ELETRÔNICAS tipo pedestal bidirecional com software gerenciador de controle-de-acesso",
        "Aquisição de medicamentos hospitalares",
        "Sistema com aplicativo facial para controle de frequência",
        "Fechaduras eletromagnéticas para porta automática",
        "Conjunto controle acesso área restrita com leitor facial e botoeira",
        "Estação de solda com controle termostático",
        "Serviço de vigilância patrimonial com controle de entrada e saída de pessoas",
    ]

    df = classificar_lista(textos)
    print(df)

    df.to_csv("saida_regex_flexivel.csv", index=False, encoding="utf-8-sig")
    print("Arquivo salvo: saida_regex_flexivel.csv")
```
